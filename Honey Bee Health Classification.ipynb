The next step is Data Augmentation. In total, the image data samples are around 5k, which isn't a lot, so I've chosen to generate __twice__ as much data, in order to reach about 10k samples. The image sample will be chosen at random and the chosen transformation will also be random, for simplicity. In this example, I have performed 4 different types of transformation (I also attempted random cropping, but the images need to be rescaled back to __100x100__, which leads to very low quality results.):
 1. __Random rotation__ - pick a random degree of rotation between 15% on the left and 15% on the right and rotate the image
 2. __Horizontal flip__ - invert the image pixels along the x axis
 3. __Vertical flip__ - invert the image pixels along the y axis
 4. __Random noise__ - add random noise to the image

```python
def random_rotation(image):
    random_degree = random.uniform(-15, 15)
    return rotate(image, random_degree)

def horizontal_flip(image):
    return image[:, ::-1]

def vertical_flip(image):
    return image[::-1, :]

def noise(image):
    return random_noise(image)

def random_crop(image):
    h = np.int(image.shape[1]/1.1)
    w = np.int(image.shape[0]/1.1)
    x = random.randint(0, image.shape[1] - h)
    y = random.randint(0, image.shape[0] - w)
    image = image[y:y+h, x:x+w]
    image = resize(image, (img_size, img_size, img_channels))
    return image

transformations = {
    'rotate': random_rotation,
    'noise': random_noise,
    'horizontal_flip': horizontal_flip,
    'vertical_flip': vertical_flip,
    #'random_crop': random_crop
}
```

The next helper-function will implement our augmentation. In order to keep the distribution of the images the same, I will generate __healthy/unhealthy bees equal to the amount of original healthy/unhealthy bees__. This means that there will be __2n__ observations for each class and __2n__ observations as a whole.
 1. Extract the data for the specified class from the dataset.
 2. If the features are already generated, just reload them, create the labels and save the data.
 3. If the features are not already generated:
     - 3.1. __Choose a random index from the data for the specified class.__  
     - 3.2. __Load the random image from disk using the file name in the dataset.__
     - 3.3. __Apply a random transformation to the image.__
     - 3.4. __Create an image path with an appropriate file name and add an appropriate label to the list of labels.__
     - 3.5. __Save the transformed image to disk.__
     - 3.6. __Append the feature to the list of features.__
     - 3.7. __Append the features list to the original features.__
     - 3.8. __Append the labels list to the original labels.__

```python
def generate_bee_images(data, feature, labels, healthy, aug_path, feat_to_append, tf_dict, num_to_generate, loaded = 0):
    
    target_bees = data[feature == healthy]
    aug_folder_path = aug_path
    features_list = []
    labels_list = []
    aug_labels = labels
    aug_features = feat_to_append
    
    if loaded == 1:
        print('Loading pre-augmented features...')
        for filename in os.listdir(aug_folder_path):
            image = load_resize_image(filename, aug_folder_path)
            if healthy:
                labels_list.append(np.int(1))
            else:
                labels_list.append(np.int(0))
            features_list.append(image)
        aug_new_features = np.array(features_list)
        aug_features = np.append(aug_features, aug_new_features, axis=0)
        aug_labels = aug_labels.append(pd.Series(labels_list))
        print('Loaded')
        return aug_features, aug_labels

    print('Creating new augmented features...')
    for i in range(num_to_generate):
        rand_idx = np.random.randint(0, len(target_bees))
        inp_file = target_bees.iloc[rand_idx]['file']
        image_to_transform = load_resize_image(inp_file)
        out_file = os.path.splitext(target_bees.iloc[rand_idx]['file'])[0]

        key = random.choice(list(tf_dict))
        #print(key)
        transformed_image = tf_dict[key](image_to_transform)
        if healthy:
            new_file_path = '%s/%s_aug_h%s.png' % (aug_folder_path, out_file, i)
            labels_list.append(np.int(1))
        else:
            new_file_path = '%s/%s_aug_u%s.png' % (aug_folder_path, out_file, i)
            labels_list.append(np.int(0))
        if not os.path.exists(aug_folder_path):
            os.makedirs(aug_folder_path)

        # write image to the disk
        imsave(new_file_path, transformed_image)
        #transformed_image = np.reshape(transformed_image, (1, img_size, img_size, img_channels))
        #bee_features_stacked = np.append(bee_features_stacked, transformed_image, axis=0)
        features_list.append(transformed_image)

    aug_new_features = np.array(features_list)
    aug_features = np.append(aug_features, aug_new_features, axis=0)
    aug_labels = aug_labels.append(pd.Series(labels_list))
    print('Loaded')
    return aug_features, aug_labels
```

### 3. Model preparation - feature extraction

Our image data still isn't ready for modelling. The features are still __4D__ arrays and need to be converted to __2D__, while keeping the relevant information. We also need to split our data into a training and testing set and apply some form of validation.

Firstly, I'll make sure that the pixel values of the features are in the range __[0,1]__.

```python
print(bee_features_aug.max(), bee_features_aug.min(), bee_features_aug.mean())
```

For feature extraction and model preparation I have chosen to apply the following steps:
 1. Transform the 4D array to 2D by combining the height, width and channel dimensions: 
 
 __output_shape = (n_samples, width \* height \* channels)__
 
 This will lead to __30000__ features (more than the observations - __curse of dimensionality__)
 
 
 2. Normalize the data to zero mean and unit variance. (The values already seem normal but I'm adding this step just in case because some models are very sensitive to unnormalized data.)
 
 3. Perform a training and test split (standard __70% / 30%__) on the normalized features and apply stratification according to the input labels (__due to unbalanced data in each class__).
 
 4. Deal with the Curse of Dimensionality by applying PCA on the training and test features (For faster model training, I'm going to try using __50__ and __200__ components).
 
 5. Create a stratified cross-validation set, which will be used in each model. (Standard __5 and 10__ fold splits.)

```python
def reduce_and_split(features, labels, pca_components, train_size=0.7, test_size=0.3, k_fold_cv_splits=5, random_state=42):
    
    features_aug_tf = features.reshape(len(features), -1)
    print('New features shape: {}'.format(features_aug_tf.shape))
    print('Scaling data to zero mean, unit variance.')
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features_aug_tf)
    attr_train, attr_test, labels_train, labels_test = train_test_split(features_scaled, 
                                                                        labels,
                                                                        train_size=train_size,
                                                                        test_size=test_size,
                                                                        stratify=labels,
                                                                        random_state=random_state)
    print('Attribute shapes: Train - {}, Test - {}'.format(attr_train.shape, attr_test.shape))
    print('Label shapes: Train - {}, Test - {}'.format(labels_train.shape, labels_test.shape))
    pca = PCA(n_components=pca_components)
    print('Performing PCA on train features..')
    attr_train_pca = pca.fit_transform(attr_train)
    print('Performing PCA on test features..')
    attr_test_pca = pca.fit_transform(attr_test)
    print('New Attribute shapes: Train - {}, Test - {}'.format(attr_train_pca.shape, attr_test_pca.shape))
    print('Creating K-fold CV split..')
    k_fold_pca = StratifiedKFold(n_splits=k_fold_cv_splits, random_state=random_state)
    print('Finished.')
    
    return attr_train_pca, attr_test_pca, labels_train, labels_test, k_fold_pca
```

Now I'll use our helper-function to create __3__ sets of data for modelling, in order to compare performance:
 1. _Original_ (non-augmented) data, __50__ features, __5__ CV folds.
 2. _Augmented_ (full) data, __50__ features, __5__ CV folds.
 3. _Augmented_ (full) data, __200__ features, __10__ CV folds.

### 4. Model training and evaluation (and hyperparameter tuning)

We're now ready to train and evaluate different models.

I'll start with a helper function for training and model optimization. It has __4__ different options:
 1. Train a baseline model with no hyperparameter tuning. (__optimizer = None__)
 2. Load a saved model from disk using `joblib` if that model has already been trained.
 3. Perform __Grid Search__ hyperparameter tuning on the model using a pre-defined grid of parameters to find the best ones by trying out all of the combinations. Also, apply the previously created __CV__ to the model and adjust the scoring function to be __F1 Score__ (good for unbalanced data).
 4. Perform __Randomized Search__ hyperparameter tuning on the modem using a pre-defined grid of parameters to try out random combinations and find the best results, according to these combinations (may lead to losing the best estimator, but is appropriate for large parameter grids). Also, apply the previously created __CV__ to the model and adjust the scoring function to be __F1 Score__ (good for unbalanced data).

When the model is defined:

 1. __Fit the model.__
 2. __Output the best found parameters and the best score.__
 3. __Display all the scores with every possible combination.__
 4. __Save the model to disk and return it.__

```python
def train_optimize_model(features_train, labels_train, model, model_name=None, cv=k_fold_pca, 
                         param_list=None, optimizer=None, 
                         num_results=10, display_scores=0):
    if optimizer is None:
        print('Training baseline model..')
        model.fit(features_train, labels_train)
        print('Training complete.')
    elif os.path.isfile(model_name):
        print('Loading saved model...')
        model = joblib.load(model_name)
        print('Loaded..')
        return model
    else: 
        if param_list is None:
                print('No parameters specified for model.')
        else:
            if optimizer == 'Grid':
                opt_model = GridSearchCV(estimator=model, param_grid=param_list, cv=cv.split(features_train, labels_train),
                                           scoring=make_scorer(f1_score, pos_label=0))
                print('Optimizing model with GridSearch...')
            if optimizer == 'Random':
                opt_model = RandomizedSearchCV(estimator=model, param_distributions=param_list, 
                                                 cv=cv.split(features_train, labels_train),
                                                 scoring=make_scorer(f1_score, pos_label=0))
                print('Optimizing model with RandomSearch...')
            opt_model.fit(features_train, labels_train)
            print('Best parameters found: {}'.format(opt_model.best_params_))
            print('Best score found: {}'.format(opt_model.best_score_))
            if display_scores:
                print('Displaying scores..')
                means = opt_model.cv_results_['mean_test_score']
                stds = opt_model.cv_results_['std_test_score']
                for mean, std, params in zip(means, stds, opt_model.cv_results_['params']):
                    print("{:.3f} ({:.03f}) for {}".format(mean, std * 2, params))
            print('Saving model to disk..')
            filename = model_name
            joblib.dump(opt_model.best_estimator_, filename)
            return opt_model
```

Now let's see what the evaluation helper-function will look like:

 1. Call predict on the training and test set to get the predicted labels for each sample.
 2. Create a __Classification Report__ (__Precision__, __Recall__, __F1 Score__, __Support__) for each set.
 3. Create a __Confusion Matrix__ on the test set and plot it.
 4. If possible (not with LinearSVC), plot a __ROC Curve__ on the test set using the predicted probabilities for the positive class (the positive class is defined as __0__ or `unhealthy`)
 
 The ROC curve plotting function calculates the __FPR__ and __TPR__, which are then used to calculate the __AUC__(Area Under the Curve) and for plotting the curve.

```python
def plot_roc(test_labels, pred_labels):
    fpr, tpr, thresholds = roc_curve(test_labels, pred_labels, pos_label=0)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    lw = 1
    plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area = {0:.2f})'.format(roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
    
def evaluate_model(features_train, labels_train, features_test, labels_test, model, 
                   class_names=['unhealthy', 'healthy'],
                   roc=True):
    print('Evaluating model..')
    labels_pred_test = model.predict(features_test)
    labels_pred_train = model.predict(features_train)
    print('Classification report for training data.')
    print(classification_report(labels_train, labels_pred_train, target_names=class_names))
    print('Classification report for test data.')
    print(classification_report(labels_test, labels_pred_test, target_names=class_names))
    print('Confusion matrix for test data.')
    cm = confusion_matrix(labels_test, labels_pred_test)
    print(cm)
    fig = plt.figure(figsize=(10,6))
    ax = fig.add_subplot(111)
    cax = ax.matshow(cm)
    plt.title('Confusion matrix')
    fig.colorbar(cax)
    ax.set_xticklabels([''] + class_names)
    ax.set_yticklabels([''] + class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
    if roc == True:
        labels_pred_prob = model.predict_proba(features_test)[:,0]
        print('Plotting ROC Curve for test data..')
        plot_roc(labels_test, labels_pred_prob)
    print('Evaluation complete.')
```

In total, I have chosen to train __4 different types of models__ - __Logistic Regression__ (baseline model), __Random Forest Classifier__ (one ensemble model) and two __SVM__ models (__Linear SVM__, and __Kernel SVM__). The main algorithms are ensembles and SVM's, which will hopefully be able to separate the data from the two classes.

### 5. __Choosing the best model and predicting new unknown data__

I will assign the path to our test images and use a helper-function to load and predict them using the two models specified above.

```python
test_folder = 'data/test_predict/'
```

The loading function is the same as for the features but with a different path and will be used to insert the images into a pandas `DataFrame`.

```python
def load_resize_image_test(file, folder=test_folder, size=img_size, num_channels=img_channels):
    img = imread(folder + file)
    img = resize(img, (size, size), mode='reflect')
    return img[:, :, :num_channels]
```

The input images need to be loaded and preprocessed in the same fashion as the training images. To do that, I must first create a `DataFrame` containing the file names and apply them to the loading function to receive the images. After that the same preprocessing steps, that were previously stated can be applied. Because the trained attributes have __50__ PCA components, I will need to predict at least __50__ images. After forming the predictions, I will create a __10__ by __5__ plot to visualize the images and their labels.

```python
def predict_test_images(model, path='data/test_predict/', pca_components=50, nrows=10, ncols=5, size=(15, 20)):
    files_list = []
    image_list = []
    for filename in os.listdir(path):
        files_list.append(filename)
        image = load_resize_image(filename, path)
        image_list.append(image)
    input_s = pd.DataFrame(files_list, columns=['input_file'])
    test_images = input_s['input_file'].apply(load_resize_image_test)
    test_images_stacked = np.stack(test_images, axis=0)
    test_features = test_images_stacked.reshape(len(test_images_stacked), -1)
    scaled_features = StandardScaler().fit_transform(test_features)
    pca_features = PCA(n_components=pca_components).fit_transform(scaled_features)
    labels_pred = model.predict(pca_features)
    print('Predicted labels for test images..')
    print(labels_pred)
    
    print('Plotting the images..')
    f, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=size)
    k = 0
    plt.title('Test predictions')
    for i in range(nrows):
        for j in range(ncols):
            ax[i,j].grid(False)
            ax[i,j].get_xaxis().set_visible(False)
            ax[i,j].get_yaxis().set_visible(False)
            ax[i,j].imshow(image_list[k])
            ax[i,j].set_title('Predicted Class:{}'.format(labels_pred[k]))
            k+=1
            
    plt.show()
```
